```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('voice.csv')
```

```python
df_corr = df.corr()
```

```python
f,ax = plt.subplots(figsize=(9,9))

sns.heatmap(df_corr,square=True)
plt.show()
```

```python
sns.set_style('dark')
df.boxplot(column='meanfreq',by='label',grid=False)
```

```python
sns.set_style('darkgrid')
df.boxplot(column='sd',by='label',grid=False)
```

```python
sns.set_style('dark')
df.boxplot(column='median',by='label',grid=False)
```

```python
#mean of the male attributes
df[df['label']=='male'].mean()
```

```python
#mean of the female attributes
df[df['label']=='female'].mean()
```

```python
sns.FacetGrid(df, hue="label", size=6).map(sns.kdeplot, "meanfun").add_legend()
```

```python
sns.FacetGrid(df,hue='label',size=6).map(sns.kdeplot,'meanfreq').add_legend()
```

```python
sns.FacetGrid(df,hue='label',size=6).map(sns.kdeplot,'IQR').add_legend()
```
